## DTW plots

```python
# time series x axis
x1 = np.linspace(0,10,101)
x2 = np.linspace(0,10,101)
# time series y axis
y1 = np.zeros(71)
y2 = np.zeros(71)
y1[10:40] = np.sin(np.linspace(0, 2*math.pi, 30))
y1 = y1 + 3
y2[20:50] = np.sin(np.linspace(0, 2*math.pi, 30))
```

```python
plt.figure(figsize=(8, 6))
plt.plot(y1, color='tab:blue', label='Time series 1')
plt.plot(y2, color='tab:orange', label='Time series 2')
for i in range(0, len(y1), 2):
    plt.plot([i, i], [y1[i], y2[i]], '-', color='0.8', linewidth=1)
    plt.plot(i, y1[i], 'o', color='tab:blue')
    plt.plot(i, y2[i], 'o', color='tab:orange')

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
ax.spines["bottom"].set_linewidth(0.6)
ax.spines["right"].set_linewidth(0.6)
ax.spines["left"].set_linewidth(0.6)
ax.spines["top"].set_linewidth(0.6)
plt.title("Euclidean distance", fontsize=20)
plt.legend(fontsize=14)
plt.savefig("euclidean.pdf", format='pdf')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(y1, color='tab:blue', label='Time series 1')
plt.plot(y2, color='tab:orange', label='Time series 2')

for j in range(0, 12, 2):
    plt.plot([0, j], [y1[0], y2[j]], '-', color='0.8', linewidth=1)

for i in range(2, len(y1) - 30, 2):
    plt.plot([i, i+10], [y1[i], y2[i+10]], '-', color='0.8', linewidth=1)
    
for j in range(40, 52, 2):
    plt.plot([j, 50], [y1[j], y2[50]], '-', color='0.8', linewidth=1)
    
for i in range(52, len(y1), 2):
    plt.plot([i, i], [y1[i], y2[i]], '-', color='0.8', linewidth=1)

for i in range(0, len(y1), 2):
    plt.plot(i, y1[i], 'o', color='tab:blue')
    plt.plot(i, y2[i], 'o', color='tab:orange')
    
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
ax.spines["bottom"].set_linewidth(0.6)
ax.spines["right"].set_linewidth(0.6)
ax.spines["left"].set_linewidth(0.6)
ax.spines["top"].set_linewidth(0.6)
plt.title("Dynamic time warping", fontsize=20)
plt.legend(fontsize=14)
plt.savefig("dtw.pdf", format='pdf')
plt.show()
```

```python
def extract_perf_task(all_task_paths: list[Path]) -> (list[float], list[float], list[float]):
    pos_error = []
    ori_error = []
    j_error = []
    for task_path in all_task_paths:
        all_repro_files = [path for path in list(Path(task_path).rglob("reproduction*"))]
        for file in all_repro_files:
            perf_file = file.joinpath("model_performance.json")
            with open(perf_file, 'r') as f:
                data = json.load(f)
            pos_error.append(data["pos_error"])
            ori_error.append(data["ori_error"])
            j_error.append(data["j_error"])
    return pos_error, ori_error, j_error
```
